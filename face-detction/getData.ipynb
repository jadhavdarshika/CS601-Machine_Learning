```python
import os
import cv2
import easygui as e
import numpy as np
from PIL import Image
import pickle
```

```python
face_dict={}
path=r"D:\Projects\face-detection"
name=""
names=[]
class generate_dataset():
    def init():
        global path
        global names
        names=e.enterbox(msg="Name of the person",title="")
        try:
            os.mkdir(path+"\\"+"dataset")
        except FileExistsError:
            pass
        
        try:
            os.mkdir(path+"\\"+"dataset"+"\\"+names)
        except:
            pass
        
        new_path=path+"\\"+"dataset"+"\\"+names
        
        total_images=int(e.enterbox("msg=Number of images for training"))
        
        cam=cv2.VideoCapture(0)
        
        print("starting the camera")
        
        detector=cv2.CascadeClassifier(r"D:\Projects\face-detection\face.xml")
        
        offset=50
        i=0
        
        while(1):
            ret,im=cam.read()
            gray=cv2.cvtColor(im,cv2.COLOR_BGR2GRAY)
            print("gray",gray)
            
            faces=detector.detectMultiScale(gray,scaleFactor=1.1,minNeighbors=5,minSize=(30,30),flags=cv2.CASCADE_SCALE_IMAGE)
            
            for(x,y,w,h) in faces:
                i=i+1
                print(i)
                print("Image Saved")
                cv2.imwrite(new_path+"\\_"+name+"."+str(i)+".jpg",gray[y-offset:y+h+offset,x-offset:x+w+offset])
                cv2.rectangle(im,(x-50,y-50),(x+w+50,y+h+50),(220,0,0),2)
                cv2.putText(im,name+" "+str(i)+"/"+str(total_images),(x-45,y+h+45),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,0,255),2)
                cv2.putText(im,"press ESC for exit",(x-45,y-45),cv2.FONT_HERSHEY_SIMPLEX,0.5,(0,255,255),2)
                print("i=",i)
            
            try:
                cv2.imshow("dataset generation start....",im)
            except:
                continue
            
            k=cv2.waitKey(100) & 0xff
            
            if k==27:
                break
            
            if i+1>total_images:
                break
        
        cam.release()
        cv2.destroyAllWindows()

generate_dataset.init()
```
